```python
from Comment_Analyze.UrunListesi_Hepsiburada import UrunListesiAlma
import pandas as pd
```

```python
base_url = input("L√ºtfen Hepsiburada √ºr√ºn listesinin URL'sini girin: ").strip()
scraper = UrunListesiAlma(base_url)

# Kullanƒ±cƒ± ka√ß sayfa √ßekeceƒüini belirleyebilir
page_limit = 40 #int(input("Ka√ß sayfa verisini almak istiyorsunuz? (Varsayƒ±lan: 5) ") or 5)
product_df = scraper.scrape(page_limit=page_limit)

# **üìå Sonu√ßlarƒ± Tabloda Yazdƒ±r**
print("\n‚úÖ √úr√ºnler Tablosu:")
print(product_df.to_string(index=False))  # Daha d√ºzenli yazdƒ±rma
```

```python
# **üìå Excel Dosyasƒ±na Kaydetme Fonksiyonu**
def save_to_excel(df, filename="urun_listesi.xlsx"):
    """Veriyi bir Excel dosyasƒ±na kaydeder."""
    try:
        df.to_excel(filename, index=False)
        print(f"\n‚úÖ Veri ba≈üarƒ±yla '{filename}' dosyasƒ±na kaydedildi!")
    except Exception as e:
        print(f"‚ö†Ô∏è Excel kaydedilirken hata olu≈ütu: {e}")
```

```python
save_to_excel(product_df, "hepsiburada_urun_listesi.xlsx")
```

```python
import pandas as pd

# CSV dosyasƒ±na kaydetme
csv_path = "product_list.csv"  # Dosya adƒ±nƒ± belirleyebilirsin
product_df.to_csv(csv_path, index=False, encoding="utf-8-sig")

print(f"‚úÖ DataFrame ba≈üarƒ±yla '{csv_path}' olarak kaydedildi!")
```
